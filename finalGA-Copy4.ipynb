```python
import random
import pandas as pd
import json

# Function to load route data from a JSON file
def load_routes_from_json(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        data = json.load(file)

    route_options = {}

    for entry in data:
        origin = entry["origin"]
        destination = entry["destination"]

        if (origin, destination) not in route_options:
            route_options[(origin, destination)] = []

        for route in entry["routes"]:
            route_summary = route["summary"]
            distance = float(route["distance"].split()[0])

            route_options[(origin, destination)].append({
                "route_name": route_summary,
                "distance": distance
            })

    return route_options

# Function to initialize population with routes
def initialize_population(csv_file, n_population, route_options):
    df = pd.read_csv(csv_file)

    POI = {city: idx for idx, city in enumerate(df['locations'])}

    print("Available cities:")
    for idx, city in enumerate(df['locations'], start=1):
        print(f"{idx}. {city}")

    start_idx = int(input("\nChoose your starting point (enter the number): ")) - 1
    start_city = df['locations'].iloc[start_idx]

    print("\nChoose the places you want to visit (enter numbers separated by commas):")
    visit_idxs = list(map(lambda x: int(x.strip()) - 1, input().split(',')))

    visit_cities = [df['locations'].iloc[i] for i in visit_idxs if i != start_idx]
    cities_to_visit = [start_city] + visit_cities

    population = []
    for _ in range(n_population):
        random_route = cities_to_visit[1:]
        random.shuffle(random_route)
        random_route = [start_city] + random_route

        route_sequence = []
        for i in range(len(random_route) - 1):
            origin = random_route[i]
            destination = random_route[i + 1]
            loc_pair = (origin, destination)

            if loc_pair not in route_options:
                loc_pair = (destination, origin)

            if loc_pair in route_options:
                shortest_route = min(route_options[loc_pair], key=lambda x: x['distance'])
                route_sequence.append(shortest_route)
            else:
                route_sequence.append(None)

        population.append({
            'route_sequence': route_sequence,
            'route_indices': [POI[city] for city in random_route]
        })

    return population

# Load route options and initialize population
file_path = "directions_results.json"
route_options = load_routes_from_json(file_path)

csv_file = "finaldata_1.csv"
n_population = 100
population = initialize_population(csv_file, n_population, route_options)
```

```python
def fitness_prob(population, route_options):
    """
    Calculate the fitness probabilities for the entire population.
    Fitness is inversely proportional to the total distance of the route.
    """
    # Calculate total distance for each individual (chromosome)
    total_dist_all_individuals = [total_dist_individual(chromosome, route_options) for chromosome in population]

    # Compute multiplicative inverse of the distances to calculate fitness
    fitness_probs = [1 / dist if dist != 0 else float('inf') for dist in total_dist_all_individuals]

    return fitness_probs

def total_dist_individual(chromosome, route_options):
    """
    Calculate the total distance for a single chromosome.
    The chromosome contains 'route_sequence' which holds the route information.
    """
    total_distance = 0
    
    # Iterate through the sequence of cities in the chromosome (excluding the last one since it's a round trip)
    for i in range(len(chromosome['route_sequence']) - 1):
        route_info = chromosome['route_sequence'][i]
        total_distance += route_info['distance']  # Add the distance of each route

    # Add the distance for the round-trip (from last city back to the start)
    last_route_info = chromosome['route_sequence'][-1]
    total_distance += last_route_info['distance']
    
    return total_distance
```
